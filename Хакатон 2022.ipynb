```python
import pandas as pd
import os
```

```python
root = "./TNVD_All"
files_all = []

for path, subdirs, files in os.walk(root):
    for name in files:
        files_all.append(os.path.join(path, name))

print(files_all)
```

```python
df_list = []
df_all = []

headers = ['napr','period','tnved','Stoim']

for file in files_all:
#     print(file)
    df_list.append(pd.read_csv(file, sep = '\t',usecols=headers,low_memory=False))

df = pd.concat(df_list)
```

```python
df.info()
```

```python
df['napr'] = df['napr'].astype(str)
df['period'] = pd.to_datetime(df['period'])
df['tnved'] = df['tnved'].astype(str)    # при необходимости, можно сделать int64

df['Stoim'] = df['Stoim'].astype(str)
df['Stoim'] = df['Stoim'].str.replace(',', ".").astype(float)

# df['Netto'] = df['Netto'].str.replace(',', ".").astype(float)
# df['Kol'] = df['Kol'].str.replace(',', ".").astype(float)

```

```python
df.info()
```

```python
df_tnvd = df.groupby(['tnved', 'napr'], as_index= False).agg({'Stoim':'sum'})
```

```python
df_tnvd.head()
```

```python
df_tnvd = df_tnvd.sort_values(by='tnved', ascending=False)
```

```python
df_tnvd.head(50)
```

```python
df2 = df_tnvd.pivot(index='tnved', columns='napr', values='Stoim')
```

```python
df2
```

```python
df2 = df2.reset_index()
```

```python
df2.head(50)
```

```python
df2['Saldo'] = df2['ЭК'] - df2['ИМ']
```

```python
df2 = df2.sort_values(by='Saldo', ascending = False)
```
